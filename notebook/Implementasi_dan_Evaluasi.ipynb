```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Load model dan tokenizer BERT2BERT Bahasa Indonesia
tokenizer = AutoTokenizer.from_pretrained("cahya/bert2bert-indonesian-summarization")
model = AutoModelForSeq2SeqLM.from_pretrained("cahya/bert2bert-indonesian-summarization")

# Fungsi ringkasan
def ringkas_teks_indo(text, max_input_length=1024, max_output_length=150):
    input_text = "summarize: " + text
    inputs = tokenizer.encode(input_text, return_tensors="pt", max_length=max_input_length, truncation=True)
    summary_ids = model.generate(
        inputs,
        max_length=max_output_length,
        min_length=40,
        length_penalty=2.0,
        num_beams=4,
        early_stopping=True
    )
    return tokenizer.decode(summary_ids[0], skip_special_tokens=True)
```

## Evaluasi

### Cosine Similarity

```python
!pip install -q scikit-learn pandas

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
manual_url = "https://docs.google.com/spreadsheets/d/1zcNv1AcX6HwV6Bsj_si15dez7PAKJWi6Up9z9Yy9XeM/export?format=csv"
manual_df = pd.read_csv(manual_url)
manual_df.columns
```

```python
# Pastikan document_id bertipe string
manual_df['document_id'] = manual_df['document_id'].astype(str)
tfidf_to_bert2bert_summary_df['document_id'] = tfidf_to_bert2bert_summary_df['document_id'].astype(str)

# Merge berdasarkan document_id
merged_df = pd.merge(tfidf_to_bert2bert_summary_df, manual_df[['document_id', 'manual_summary']], on='document_id', how='inner')
```

```python
def get_similarity(row):
    texts = [str(row['tfidf_to_bert2bert_summary']), str(row['manual_summary'])]
    tfidf = TfidfVectorizer().fit_transform(texts)
    return cosine_similarity(tfidf[0:1], tfidf[1:2])[0][0]

merged_df['cosine_similarity'] = merged_df.apply(get_similarity, axis=1)
```

```python
merged_df[['document_id', 'tfidf_to_bert2bert_summary', 'manual_summary', 'cosine_similarity']].head(50)
```

```python
average_cosine = merged_df['cosine_similarity'].mean()
average_cosine = average_cosine*100
print(f"Average Cosine Similarity: {average_cosine:.0f}%")
```

### BERTScore

```python
!pip install bert-score

from bert_score import score
```

```python
# Ambil daftar teks
candidates = merged_df['tfidf_to_bert2bert_summary'].astype(str).tolist()
references = merged_df['manual_summary'].astype(str).tolist()

# Hitung BERTScore
P, R, F1 = score(candidates, references, lang="id")

# Simpan ke kolom baru
merged_df['bertscore_precision'] = [p.item() for p in P]
merged_df['bertscore_recall'] = [r.item() for r in R]
merged_df['bertscore_f1'] = [f.item() for f in F1]
```

```python
merged_df[['document_id', 'tfidf_to_bert2bert_summary', 'manual_summary', 'bertscore_f1']].head(50)
```

```python
# Hitung rata-rata dari masing-masing metrik
avg_precision = merged_df['bertscore_precision'].mean()*100
avg_recall = merged_df['bertscore_recall'].mean()*100
avg_f1 = merged_df['bertscore_f1'].mean()*100

# Cetak hasil
print(f"Average BERTScore Precision : {avg_precision:.0f}%")
print(f"Average BERTScore Recall    : {avg_recall:.0f}%")
print(f"Average BERTScore F1        : {avg_f1:.0f}%")
```
